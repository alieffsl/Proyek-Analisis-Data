### Menggabungkan dua dataset berdasarkan kolom "dteday" 

```python
def merge_df(df_day, df_hour):
  df_day = df_day.copy()
  df_hour = df_hour.copy()

  day_feature = ['temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered', 'cnt']
  day_feature_dict = {x:f"{x}_day" for x in day_feature}
  df_day.rename(columns=day_feature_dict, inplace=True)

  drop_measure = [x for x in df_day.columns if x not in day_feature_dict.values()]
  drop_measure.remove('dteday')
  df_day.drop(drop_measure, axis=1, inplace=True)

  df_merge = pd.DataFrame(df_hour.merge(df_day, on="dteday", how="left"))

  return df_merge
```

```python
df_merge = merge_df(df_day, df_hour)
df_merge.head()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana trend penyewaan sepeda di setiap bulan?

```python
to_month = {
    1: "January",
    2: "February",
    3: "March",
    4: "April",
    5: "May",
    6: "June",
    7: "July",
    8: "August",
    9: "September",
    10: "October",
    11: "November",
    12: "December"
}

to_year = {
    0: 2011,
    1: 2012
}

def plot_by_month(df: pd.DataFrame, feature='cnt', return_df=False):
    df = df.copy()
    df.sort_values(['yr', 'mnth'], inplace=True)
    df['month'] = df['mnth'].apply(lambda x: to_month[x])
    df['year'] = df['yr'].apply(lambda x: str(to_year[x]))
    df['month_year'] = df['month'] + " " + df['year']

    df_feature_col = df.groupby('month_year').mean(numeric_only=True)[feature]
    df[f'month_{feature}'] = df['month_year'].apply(lambda x: df_feature_col[x])
    df_result = df[['month_year', f'month_{feature}']].set_index('month_year')

    if return_df:
        return df_result

    plt.figure(figsize=(8, 5))
    sns.barplot(x=f'month_{feature}', y='month_year', data=df_result, orient='h')
    plt.title(f'Monthly Trend of Bike Rents by {feature}')
    plt.show()
```

```python
plot_by_month(df_day, 'cnt')
```

### Pertanyaan 2: Pada Jam berapa puncak penyewaan sepeda terjadi?

```python
hr_format = {
    0: '12 am', 1: '1 am', 2: '2 am', 3: '3 am',
    4: '4 am', 5: '5 am', 6: '6 am', 7: '7 am',
    8: '8 am', 9: '9 am', 10: '10 am', 11: '11 am',
    12: '12 pm', 13: '1 pm', 14: '2 pm', 15: '3 pm',
    16: '4 pm', 17: '5 pm', 18: '6 pm', 19: '7 pm',
    20: '8 pm', 21: '9 pm', 22: '10 pm', 23: '11 pm'
}

def plot_by_hour(df: pd.DataFrame, col='cnt'):
    df = df.copy()
    df_hr_col = df.groupby('hr').mean(numeric_only=True)[col]

    plt.figure(figsize=(15, 6))
    sns.barplot(x=df_hr_col.index.map(hr_format), y=df_hr_col)
    plt.title(f'Hourly Trend of {col.capitalize()}')
    plt.xlabel('Hour of the Day')
    plt.ylabel(col.capitalize())
    plt.show()
```

```python
plot_by_hour(df_merge)
```

### Pertanyaan 3: Apa pengaruh musim terhadap tingkat ketersewaan sepeda?

```python
season = {
    1: 'Spring',
    2: 'Summer',
    3: 'Fall',
    4: 'Winter'
}

def plot_by_season(df: pd.DataFrame, col='cnt'):
    df = df.copy()
    df_season_col = df.groupby('season').mean(numeric_only=True)[col]

    plt.figure(figsize=(5, 5))
    plt.pie(df_season_col, labels=df_season_col.index.map(season), autopct='%1.1f%%', startangle=140)
    plt.title('Ketersewaan Sepeda Berdasarkan Musim')
    plt.show()
```

```python
plot_by_season(df_merge)
```
